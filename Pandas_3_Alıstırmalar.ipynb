5) Arabanın modeli 75-80 yılları arasındaki arabaların hepsini getirin.(75 ve 80 dahil)


```python
df[(df['model_year'] >= 75) & (df['model_year'] <= 80)]
```

6) horsepower'ı 155-180 arasındaki arabaların hepsini getirin.(155 dahil, 180 dahil değil)(Bu sorguyu yapabilmeniz için horsepower sütununu object durumdan float duruma getirmeniz gerekiyor.)





```python
df['horsepower'].replace('?' , np.nan , inplace = True)
df['horsepower'] = df['horsepower'].dropna().astype(float)
df[(df['horsepower']<180) & (df['horsepower']>=155)]
```

7) Elinizdeki df'i horsepower'a göre büyükten küçüğe sıralayın ardından df'in indexlerini sıfırlayın.


```python
df = df.sort_values(by='horsepower', ascending=False)
df = df.reset_index(drop=True)
df
```

15) Veriyi istediğiniz şekilde inceleyerek bu zamana kadar öğrendiğiniz bilgilerle anlamlı 5 adet yargı cümlesi çıkartın. ( En çok insan Asya kıtasında yaşıyor, En çok insan Afrinın sahralarında ölüyor, Ortalama en yüksek telefon sahibi Kuzey Amerikada, En çok okuma yazma oranına sahip ülkeler Baltık ülkeleri vb.)

```python
#Kişi başına düşen milli gelire sahip ülke
df_new[df_new["GDP ($ per capita)"] == df_new["GDP ($ per capita)"].min()]["Country"].iloc[0]
```

```python
#"SUB-SAHARAN AFRICA" bölgesindeki ölüm oranlarının ortalaması
deaths = df_new[df_new.Region == 'SUB-SAHARAN AFRICA']['Deathrate'].mean()
deaths
```

```python
#En yüksek doğum oranına sahip ülke
max_birthrate = df_new["Birthrate"].max()
country_with_max_birthrate = df_new.loc[df_new["Birthrate"] == max_birthrate]["Country"].iloc[0]
print(f"En yüksek doğum oranına sahip ülke: {country_with_max_birthrate}")
```

```python
#En çok okuma yazma oranına sahip ülkeler
max_literacy_rate = df_new["Literacy (%)"].max()
countries_with_max_literacy = df_new.loc[df_new["Literacy (%)"] == max_literacy_rate]["Country"]
countries_with_max_literacy
```

```python
#Kuzey Amerika ülkeleri arasında ortalama en yüksek telefon kullanımı
max_phone_owners = df_new[df_new.Region == "NORTHERN AMERICA"]["Phones (per 1000)"].max()
max_phone_owners
```
